Importing All Dependencies

```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns

%matplotlib inline
```

* Loading Datasets

```python
data = pd.read_csv('airbnb dataset.csv', encoding_errors='ignore')
```

* Data Cleaning

```python
data.isnull().sum()

# Dropping missing values

data.dropna(inplace=True)
```

```python
data.shape
```

```python
# Dealing with Duplicates

data.duplicated().sum()

data[data.duplicated()]

#Deleting Duplicates

data.drop_duplicates(inplace=True)
```

```python
data.duplicated().sum()
```

### UNIVARIATE ANALYSIS

```python
# Identifing Outliers
plt.figure(figsize=(10,4))
sns.boxplot(data=data, x = 'price')
plt.title("Boxplot of price")
plt.show()
```

```python
# Filtered data to exclude Outliers

df=data[data['price']<1500]
```

```python
plt.figure(figsize=(10,5))
sns.histplot(data=df , x ='price',bins=30,kde=True)
plt.title("histogram of price")
plt.xlabel("price")
plt.ylabel("frequency")
plt.show()
```

```python
plt.figure(figsize=(10,5))
sns.histplot(data=df , x ='availability_365',bins=30,kde=True)
plt.title("availability_365 distribution")
plt.xlabel("price")
plt.ylabel("frequency")
plt.show()
```

```python
df.groupby(by='neighbourhood_group')['price'].mean()
```

# Feature Engineering

```python
# Creating a new column

df["price per bed"] = df['price']/df['beds']
df.head()
```

```python
# Average price per bed

df.groupby(by='neighbourhood_group')['price per bed'].mean()
```

# Bivariate Analysis

```python
df.columns
```

```python
# Price dependency on neighbour hood

sns.barplot(data=df,x='neighbourhood_group',y='price',hue='room_type')

plt.show()
```

```python
# Relation between price and no.of reviews
plt.figure(figsize=(10,5))
sns.scatterplot(data=df,x='number_of_reviews',y='price',hue='neighbourhood_group')
plt.title("reviews and locality analysis")
plt.show()
```

```python
sns.pairplot(data=df,vars=['price','minimum_nights','number_of_reviews','availability_365'],hue='room_type')
plt.show()
```

# Correlation analysis

```python
corr = df[['price','minimum_nights','number_of_reviews','reviews_per_month','availability_365','beds']].corr()
corr
```

```python
plt.figure(figsize=(8,6))
sns.heatmap(data=corr,annot=True)
plt.show()
```
